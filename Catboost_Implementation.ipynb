```python
# catboost classifier

from sklearn.datasets import make_classification
from catboost import CatBoostClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
x,y = make_classification(n_samples=1000,n_features=20,n_classes=2,random_state=1)
```

```python
x
```

```python
y
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
clf = CatBoostClassifier(verbose = 0) # if you dont want details in training put verbose = 0
clf
```

```python
clf.fit(x_train,y_train)
```

```python
y_pred = clf.predict(x_test)
y_pred
```

```python
# evaluation of the model

print('Current model performance:')
print(f'Accuracy:{accuracy_score(y_test,y_pred)}')
print('Classification Report:')
print(classification_report(y_test,y_pred))
print('Confusion Matrix:')
print(confusion_matrix(y_test,y_pred))
```

```python
# Hyperparameter tuning

from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {'iterations':[100,200,300],'learning_rate':[0.01,0.05,0.1,0.2],'depth':[3,4,5,6]}
```

```python

grid_search = GridSearchCV(estimator= clf,param_grid=param_grid,cv=5,verbose =3)
```

```python
grid_search
```

```python
grid_search.fit(x_train,y_train)
```

```python
grid_search.best_params_
```

```python
best_model = grid_search.best_estimator_
best_model
```

```python
y_pred = best_model.predict(x_test)
y_pred

# evaluation of the model

print('Current model performance:')
print(f'Accuracy:{accuracy_score(y_test,y_pred)}')
print('Classification Report:')
print(classification_report(y_test,y_pred))
print('Confusion Matrix:')
print(confusion_matrix(y_test,y_pred))
```

#Catboost Regressor

```python

from sklearn.datasets import make_regression
from catboost import CatBoostRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
x,y = make_regression(n_samples=1000,n_features=2,noise=10,random_state=1)
```

```python
x
```

```python
y
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
regressor =  CatBoostRegressor()
regressor
```

```python
regressor.fit(x_train,y_train)
```

```python
y_pred = regressor.predict(x_test)
y_pred
```

```python
# Evaluation of the model


from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error,mean_absolute_error

print('Current model performance:')
print(f'Accuracy:{r2_score(y_test,y_pred)}')
print(f'Mean Squared Error:{mean_squared_error(y_test,y_pred)}')
print(f'Mean Absolute Error:{mean_absolute_error(y_test,y_pred)}')
```

```python
# Hyperparameter tuning

from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {'n_estimators':[50,100,200],'learning_rate':[0.01,0.1,1.0,1.5,2.0]}
```

```python
grid_search = GridSearchCV(estimator=regressor,param_grid=param_grid,cv=5,verbose=3)
grid_search
```

```python
grid_search.fit(x_train,y_train)
```

```python
best_model = grid_search.best_estimator_
best_model
```

```python
y_pred = best_model.predict(x_test)
y_pred

# Evaluation of the model

print('Current model performance:')
print(f'Accuracy:{r2_score(y_test,y_pred)}')
print(f'Mean Squared Error:{mean_squared_error(y_test,y_pred)}')
print(f'Mean Absolute Error:{mean_absolute_error(y_test,y_pred)}')
```
